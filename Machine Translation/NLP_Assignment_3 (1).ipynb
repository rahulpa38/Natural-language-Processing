```python
## Preprocess Text
def preprocess_text(text):
  ## Lower the text
  text = text.lower()
  text_punct = ''
  ## Remove Punctuation from text
  for str in text:
    for character in str:
      if character not in punctuation:
        text_punct += character
      else:
        text_punct += ''
  ##   replace '\n'
  text_punct = text_punct.replace('\n',' ')
  return text_punct
  
```

```python
import nltk
from nltk import tokenize

nltk.download('punkt')
def preprocess_for_sentence(text):
  ## Lower the text
  text = text.lower()
  text_punct = ''
  ## Remove Punctuation 
  for str in text:
    for character in str:
      if character not in punctuation:
        text_punct += character
      else:
        text_punct += ''
  return text_punct



## Laanguage Sentences preprocessing
english_sentence_text = preprocess_for_sentence(english_text)
french_sentence_text = preprocess_for_sentence(french_text)
english_german_sentence_text = preprocess_for_sentence(english_german_text)
german_sentence_text = preprocess_for_sentence(german_text)
  
  
## language sentence array  
english_sentences = english_sentence_text.split('\n')
french_sentences = french_sentence_text.split('\n')
german_sentences = german_sentence_text.split('\n')
english_german_sentences = english_german_sentence_text.split('\n')

  
```

```python
## Split texts
words_english = english_text.split()
words_french = french_text.split()
words_german = german_text.split()
words_english_german = english_german_text.split()
```

```python
## get the vocabulary
vocab_english = set(words_english)
unique_english_words = len(vocab_english)
vocab_french = set(words_french)
vocab_german = set(words_german)
vocab_english_german = set(words_english_german)
unique_english_german_words = len(vocab_english_german)
```

```python
# vocab_french
```

```python
init_eng_trans_prob = {}

## Initialize the probabilities of english in english vs french part
for eng_word in vocab_english:
  init_french_trans_prob = {}
  for french_word in vocab_french:
     init_french_trans_prob[french_word] = (1/unique_english_words)
  init_eng_trans_prob[eng_word] = init_french_trans_prob
```

```python
## German English Initialization

init_eng_ger_trans_prob = {}

## Initialize the probabilities of english in english vs german part
for eng_word in vocab_english_german:
  init_german_trans_prob = {}
  for german_word in vocab_german:
     init_german_trans_prob[german_word] = (1/unique_english_german_words)
  init_eng_ger_trans_prob[eng_word] = init_german_trans_prob
```

```python
from copy import deepcopy
import time 

def word_translation_probabilities(lang):
  
## Initialize variables for german language  
  if lang == 'german':
    lv_vocab_english = vocab_english_german
    lv_vocab_other  = vocab_german
    init_trans_prob_eng = init_eng_ger_trans_prob
    lv_english_sentences = english_german_sentences
    lv_other_lang_sentences = german_sentences
## Initialize variables for french language    
  if lang == 'french':
    lv_vocab_english = vocab_english
    lv_vocab_other  = vocab_french
    init_trans_prob_eng = init_eng_trans_prob
    lv_english_sentences = english_sentences
    lv_other_lang_sentences = french_sentences    
      
## convergence point    
  convergence_point = 0.001  
  
## Initialize variables  
  total_f = {}
  sum_of_difference_in_prob = 999999
  trans_prob = deepcopy(init_trans_prob_eng)
  iter = 0
  
## loop for convergence  
  while (sum_of_difference_in_prob > convergence_point):
    
## Iteration time     
    iteration_start_time = time.time()
    
## Copy trans prob to previous trans probabilities    
    previous_translation_probabilities = deepcopy(trans_prob)
    
    ## initialize total(f)
    for other_lang_word in lv_vocab_other:
      total_f[other_lang_word] = 0
      
          
    ## initialize count(e,f) to 0
    counts = {}
    for english_word in lv_vocab_english:
      initialize_other_lang = {}
      for other_lang_word in lv_vocab_other:
        initialize_other_lang[other_lang_word] = 0
      counts[english_word] = initialize_other_lang 
     
    ## sentence pairs
    for english_sentence,other_lang_sentence in zip(lv_english_sentences,lv_other_lang_sentences):
      english_words = english_sentence.split()
      other_lang_words = other_lang_sentence.split()
      
      sum = {}
      ##set sum = 0 
      for english_word in english_words:
        sum[english_word] = 0
        for other_lang_word in other_lang_words:
          sum[english_word] += trans_prob[english_word][other_lang_word]
      
      ## set counts and total of alignment probabilites 
      for english_word in english_words:
        for other_lang_word in other_lang_words:
          counts[english_word][other_lang_word] += trans_prob[english_word][other_lang_word]/sum[english_word]                                       
          total_f[other_lang_word] += trans_prob[english_word][other_lang_word]/sum[english_word]
                       
#     update translation probabilities            
    for other_lang_word in lv_vocab_other:
      for english_word in lv_vocab_english:
        trans_prob[english_word][other_lang_word] = counts[english_word][other_lang_word]/total_f[other_lang_word]
        
## checking convergence
# lets seperate english and french word and perform a loop to compare their values
    sum_of_difference_in_prob = 0
    for(english_word,other_lang_word) in zip(lv_vocab_english,lv_vocab_other):
      difference_in_prob = (trans_prob[english_word][other_lang_word] - previous_translation_probabilities[english_word][other_lang_word])**2
      sum_of_difference_in_prob += difference_in_prob
    sum_of_difference_in_prob = sum_of_difference_in_prob**0.5
    
    iter = iter + 1 
    
    print("Iteration - {} Time {} - Error - {}".format(iter , time.time() - iteration_start_time , sum_of_difference_in_prob) )
  return trans_prob
      
```

```python
start = time.time()
french_trans_prob = word_translation_probabilities('french')
print("Execution Time ",time.time() - start)
```

```python
start = time.time()
german_trans_prob = word_translation_probabilities('german')
print("Execution Time ",time.time() - start)
```
