```python
import nltk
from nltk.corpus import udhr
import string
import re
english = udhr.raw('English-Latin1')
french = udhr.raw('French_Francais-Latin1')
italian = udhr.raw('Italian_Italiano-Latin1')
spanish = udhr.raw('Spanish_Espanol-Latin1')
```

```python
english_train, english_dev = english[0:1000], english[1000:1100]
french_train, french_dev = french[0:1000], french[1000:1100]
italian_train, italian_dev = italian[0:1000], italian[1000:1100]
spanish_train, spanish_dev = spanish[0:1000], spanish[1000:1100]
```

```python
english_test = udhr.words('English-Latin1')[0:1000]
french_test = udhr.words('French_Francais-Latin1')[0:1000]
italian_test = udhr.words('Italian_Italiano-Latin1')[0:1000]
spanish_test = udhr.words('Spanish_Espanol-Latin1')[0:1000]
```

```python
def preprocess_data(data,type_of_data):
## if data is train - remove punctuation(symbols and \n) from training data
    if type_of_data == 'train':
        data1 = data.lower()
        data2 = data1.replace('\n','')
        text_punct = [''.join(c for c in str if c not in string.punctuation) for str in data2]
        return text_punct

## if data is test - Lowercase the data
    if type_of_data == 'test':
        text_lower = [word.lower() for word in data]
        return text_lower
```

```python
english_train = preprocess_data(english_train,'train')
english_dev = preprocess_data(english_dev,'train')
english_test = preprocess_data(english_test,'test')

## Length of english test data
len_english = len(english_test)

french_train = preprocess_data(french_train,'train')
french_dev = preprocess_data(french_dev,'train')
french_test = preprocess_data(french_test,'test')

## Length of french test data
len_french = len(french_test)

italian_train = preprocess_data(italian_train,'train')
italian_dev = preprocess_data(italian_dev,'train')
italian_test = preprocess_data(italian_test,'test')

## Length of italian test data
len_italian = len(italian_test)

spanish_train = preprocess_data(spanish_train,'train')
spanish_dev = preprocess_data(spanish_dev,'train')
spanish_test = preprocess_data(spanish_test,'test')

## Length of spanish test data
len_spanish = len(spanish_test)
```

```python
def gram_model(data,n_gram):

## Calculate frequnce distribution of Training data
    data_unigram = nltk.FreqDist(nltk.ngrams(data,n_gram))
    return data_unigram
```

```python
## Unigram Frequency Distribution of data

english_unigram_train = gram_model(english_train,1)
english_unigram_dev = gram_model(english_dev,1)

french_unigram_train = gram_model(french_train,1)
french_unigram_dev = gram_model(french_dev,1)

italian_unigram_train = gram_model(italian_train,1)
italian_unigram_dev = gram_model(italian_dev,1)

spanish_unigram_train = gram_model(spanish_train,1)
spanish_unigram_dev = gram_model(spanish_dev,1)
```

```python
## Bigram Frequency Distribution of data

english_bigram_train = gram_model(english_train,2)
english_bigram_dev = gram_model(english_dev,2)

french_bigram_train = gram_model(french_train,2)
french_bigram_dev = gram_model(french_dev,2)

italian_bigram_train = gram_model(italian_train,2)
italian_bigram_dev = gram_model(italian_dev,2)


spanish_bigram_train = gram_model(spanish_train,2)
spanish_bigram_dev = gram_model(spanish_dev,2)
```

```python
##Getting Freq Dist of trigram model 

english_trigram_train = gram_model(english_train,3)
english_trigram_dev = gram_model(english_dev,3)

french_trigram_train = gram_model(french_train,3)
french_trigram_dev = gram_model(french_dev,3)

italian_trigram_train = gram_model(italian_train,3)
italian_trigram_dev = gram_model(italian_dev,3)

spanish_trigram_train = gram_model(spanish_train,3)
spanish_trigram_dev = gram_model(spanish_dev,3)
```
